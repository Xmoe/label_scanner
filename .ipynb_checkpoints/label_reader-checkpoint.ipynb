```python
import cv2 as cv
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
from math import sqrt

mpl.rcParams['figure.dpi'] = 160 # setting the size of the inline display image

image = cv.imread("sample.png")
#image = cv.imread("sample_external.jpg")
output = image.copy()


def show_image(image, title=""):
    plt.imshow(image, cmap="Greys_r")
    plt.title(title)
    plt.show()

#print(cv.__version__)    
    
show_image(image)
```

```python
channels = cv.text.computeNMChannels(image)
# Append negative channels to detect ER- (bright regions over dark background)
cn = len(channels)-1
for c in range(0,cn):
  channels.append((255-channels[c]))

corners = []

for channel in channels:
    er_classifier1 = cv.text.loadClassifierNM1("trained_classifierNM1.xml")
    er_filter1 = cv.text.createERFilterNM1(er_classifier1, thresholdDelta=16, minArea=0.00015, maxArea=0.20, minProbability=0.3, nonMaxSuppression=True, minProbabilityDiff=0.1)

    er_classifier2 = cv.text.loadClassifierNM2("trained_classifierNM2.xml")
    er_filter2 = cv.text.createERFilterNM2(er_classifier2, minProbability=0.8)
    
    regions = cv.text.detectRegions(channel, er_filter1, er_filter2)
    
    rects = cv.text.erGrouping(image, channel, [region.tolist() for region in regions])
    
    for r in range (0, np.shape(rects)[0]):
        rect = rects[r]
        cv.rectangle(output, (rect[0], rect[1]), (rect[0]+rect[2],rect[1]+rect[3]), (0, 0, 0), 2)
        cv.rectangle(output, (rect[0], rect[1]), (rect[0]+rect[2],rect[1]+rect[3]), (0, 255, 0), 1)
        cv.rectangle(output, (rect[0]+1, rect[1]+rect[3]-1), (rect[0]+1, rect[1]+rect[3]+1), (255, 0, 0), 2)
        
        lower_left_corner = (rect[0], rect[1]+rect[3])
        #print(f"Lower left corner found at: {lower_left_corner}")
        corners.append(lower_left_corner)
   
"""

def filter_similar(corners):
    to_filter = corners
    filtered_corners = []
    index = 0
    while True:
        try:
            current_corner = to_filter[index]
        except IndexError:
            break
        filtered_corners = [point for point in to_filter if not _are_similar(current_corner, point)]
        index += 1
        print(filtered_corners)
        to_filter = filtered_corners
    return filtered_corners
    
def _are_similar(p1, p2, threshold = 3):
    euclidean_distance = sqrt( (p1[0]-p2[0])**2 + (p1[1]-p2[1])**2 )
    return euclidean_distance <= threshold
"""


corners = list(set(corners))
corners.sort()
for corner in corners:
    print(corner)
    
        
show_image(output)
```

```python
# line detection

output = image.copy()
grey = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
kernel = np.ones((3,3),np.uint8)
erosion = cv.dilate(grey,kernel,iterations = 1)
edges = cv.Canny(erosion,50,150,apertureSize = 3)
lines = cv.HoughLines(edges,1,np.pi/180,200)
for line in lines:
    rho,theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))
    cv.line(output,(x1,y1),(x2,y2),(0,0,255),2)

show_image(erosion)
show_image(output)
```
